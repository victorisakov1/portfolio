```python
# importing libraries
import pandas as pd                   # data science essentials
import matplotlib.pyplot as plt       # essential graphical output
import seaborn as sns                 # enhanced graphical output
import numpy as np                    # mathematical essentials
import statsmodels.formula.api as smf # regression modeling


# setting pandas print options
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)


# specifying file path
file = './datasets/Ames_Housing_Dataset.xlsx'


# reading the file into Python
housing = pd.read_excel(io     = file,
                        header = 0   )

#first 50 rows
housing.head(50)
```

```python
#dropping Categorical variables for now

housing = housing.drop(["Street", "Lot_Config", "Neighborhood"], axis = 1)
```

<h3>Step 1: Hypothesize on Features to Engineer</h3><br>
<strong>a)</strong> Write a hypothesis on the effect you believe each new feature will have on the response variables (<em>Sale_Price</em> AND <em>log_Sale_Price</em>) in a markdown cell. (minimum 3 sentences per feature)

<strong>b)</strong> Give a practical, real world explanation as to why do you feel the new feature adds value. (minimum of 2 sentences per feature)

1) First Feature: Total_Covered_Area - when we combine all the area (SF) available that comes with the house - we believe the price will increase. It should be a strong positive correlation as logically, the more area/space available for the customer - the higher the Sale_Price should be (at least in general). Also, log_Sale_Price is supposed to have a bit stronger affect due to its quality of Sale_Price transformation. And when the Total_Covered_Area is smaller, the Sale_Price will be less as well.

- Practical example: When customer looking to buy a house - they want to understand what will they be able to fit there: a family of 2 or 5, 1 or 3 cars, will they have a proper Second floor, does it come with a small or big basement. Overall, it provides clients a better overview of the size of the amenities they are paying for, and should significantly affect people's decisions, especially when they buy properties. It simply explains why customers will pay less or more, as the land on it's own is expensive. So thinking of it as buying X chunk of area should be helpful. 

2) Second Feature: house_quality_condition is where we multiply the overall condition of the house and the overall quality of the house, and the divide the figure with 10, to get a relationship factor on a scale of 1 to 10. The goal is to combine the quality of the material used to build the house and the overall condition, which will positively correlate with the price of the house. We believe that a house with a high material quality and condition relationship will attract a higher price.

- Practical example: When a real estate customer is looking for a house, they not only care about the current condition of the house, they would care about the material used in the house quite significantly. - Practical example: The inclusion of a fireplace in a house, whether incorporated during the initial construction or added later, typically complicates the building process and involves additional investment. <h3>Step 2: Code the New Features</h3><br>
Use the code cell below to develop the features you have hypothesized.

```python
## Feature 1 ##

#Total Covered Area = First_Flr_SF + Second_Flr_SF + Total_Bsmt_SF + Garage_Area
housing['Total_Covered_Area'] = housing['First_Flr_SF'] + housing['Second_Flr_SF'] + housing['Total_Bsmt_SF'] + housing['Garage_Area']
```

<br>

```python
## Feature 2 ##

#Condition_of_quality = (Overall_Qual * Overall_Cond) / 10
housing['house_quality_condition'] = (housing['Overall_Qual'] * housing['Overall_Cond']) / 10
```

<br>

```python
## Feature 3 ##

#has_Fireplace = The initial feature "Fireplace" will be organized within a table to clearly indicate the presence of a fireplace in a house. The table will consist of two columns labeled "No" and "Yes," corresponding to houses without and with a fireplace, respectively.

#Checking whether both columns ("No" and "Yes") has a count of more then 100. If so, we can engineer the feature :)
Fireplace_zeroes   = len(housing['Fireplaces'][housing['Fireplaces'] == 0])

#printing the results
print(f"""
                 No\t\tYes
               ---------------------
Basement       | {Fireplace_zeroes}\t\t{len(housing) - Fireplace_zeroes}
""")

#engineering the feature
housing['has_Fireplace'] = 0


for index, value in housing.iterrows():
    
    if housing.loc[index, 'Fireplaces'] > 0:
        housing.loc[index, 'has_Fireplace'] = 1

housing.head(n=50)
```

<h3>Step 3: Check the Results</h3><br>
Develop a correlation matrix or a heatmap to show the linear relationships between your three new features and the response variables (<em>Sale_Price</em> AND <em>log_Sale_Price</em>).

```python
# code here


features= ['has_Fireplace', 'house_quality_condition', 'Total_Covered_Area', 'Sale_Price', 'log_Sale_Price']

# Calculating the correlation matrix
corr_matrix = housing[features].corr()

corr_matrix.loc[ ['has_Fireplace', 'house_quality_condition', 'Total_Covered_Area'],
              ['Sale_Price', 'log_Sale_Price'] ].sort_values(by = 'Sale_Price',
                                                             ascending = False)
```
