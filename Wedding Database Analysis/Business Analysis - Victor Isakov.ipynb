```python
#Citation: https://www.pythonfordatascience.org/chi-square-test-of-independence-python/

# setting the contigency table
cont_table = pd.crosstab(vendors_info['vendor_depart'], vendors_info['vendor_sustainable'])

# performing the chi-squared test, finding p-value, degrees of freedom, and expected frequency table
chi2, p_value, dof, expected_freq = chi2_contingency(cont_table)

# displaying the results
print("Contingency Table:\n\n", contingency_table)
print("\nChi-Squared Statistic:", chi2.round(decimals=2))
print("P-value:", p_value)
print("Degrees of Freedom:", dof)
print("\nExpected Frequencies:\n", expected_freq.round(decimals=2))
```

Table above represents how many products each department sells that either have sustainable practices(index of 1) or don't (index of 0). This chi-squared test was performed to identify whether vendors are sustainable or not. 

Also, it shows expected/predicted values for each departments, and in some departments it differs from the predictions.
For example, Hair and Makeup department was predicted to have more sustainable products, but it is the opposite. Which states that in order to understand the business question deeper, there is a need to discuss each department's performance separately. Which requires potentially more samples, but we will go into other factors that help us answer business question.

Finally, low p-value shows a statistical significance that sustainability varies across different vendors.

```python
#Citation: https://pandas.pydata.org/pandas-docs/version/0.25.0/reference/api/pandas.core.groupby.GroupBy.mean.html


# filter out rows where price_unit, price_ce, and vendor_rating is 0 or NaN
filtered_vendors_info = vendors_info[
    (vendors_info['price_unit'] != 0) & 
    (vendors_info['price_ce'] != 0) & 
    (vendors_info['vendor_rating'] != 0)
].dropna(subset=['price_unit', 'price_ce', 'vendor_rating'])

# calculating mean values
mean_values = filtered_vendors_info.groupby('vendor_sustainable')[['price_unit', 'price_ce', 'vendor_rating']].mean().round(decimals=2)

# counting unique vendors based on vendor_id for both sustainable and not sustainable practices
unique_vend_sust = filtered_vendors_info[filtered_vendors_info['vendor_sustainable'] == 1]['vendor_id'].nunique()
unique_vend_not_sust = filtered_vendors_info[filtered_vendors_info['vendor_sustainable'] == 0]['vendor_id'].nunique()

# adding the count of unique vendors to the mean_values
mean_values['unique_vendors_count'] = [unique_vend_not_sust, unique_vend_sust]

mean_values
```

The table above presents mean results of price per unit, price index, vendor rating, and the number of unique vendors that are sustainable.


The following pattern can be established:

Vendors with sustainable practices have higher price and price index, and at the same time they have slightly lower customer satisfaction (rating) which supports the idea that sustainable vendors are not bringing a higher value to customers.

Finding the means helps to understand that potentially, Vendors with No Sustainable practices bring same value for the lower amount of money which suggests that Sustainable Vendors are not more cost effective.

```python
# importing a package for Regression Model
import statsmodels.formula.api as smf

# setting values that will be in the model
lm_best = smf.ols(formula = """vendor_rating ~ vendor_sustainable+
                                          price_ce""",
                   data    = filtered_vendors_info)


# fitting the data into the model
results = lm_best.fit()


# creating summary
print(results.summary())
```

The OLS Regression model explains only 3.7% of the relationship.

F-statistic suggests that the model is statistically significant and relationship of sustainability on price index is not random.
It's P-value is below 0.05 which indicates statistical significance. 

P-value for vendor sustainability index to rating is below 0.05 which suggests that there is a statistically significant relationship.
P-value for vendor rating and price index is much higher than 0.05 which suggests there might be a random association rather than a real effect.

Finally, this OLS Model suggests there is no significant relation of price index to rating, so we need to do another relationship model analysis in the next step.

```python
# importing a package for Regression Model
import statsmodels.formula.api as smf

# setting values that will be in the model
lm_best = smf.ols(formula = """ price_unit ~ vendor_rating
                                         """,
                   data    = filtered_vendors_info)


# fitting the data into the model
results = lm_best.fit()


# creating summary
print(results.summary())
```

This model suggests there is a statistically significant relationship between price per unit and vendor rating.
It suggests that as price increases, the rating decreases.

It supports the idea that more expensive products provided by vendors give less cost effectiveness to the customer.
This final OLS model finalized a weak but significant relation of decrease of cost effectiveness with price increase.
